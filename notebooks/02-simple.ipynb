```python
from typing import TypedDict

class State(TypedDict):
    customer_name: str
    my_age: int

state: State = {}
customer_name = state.get("customer_name", None)
print(customer_name)
```

```python
#Definimos un nodo como una función
#Todos los nodos tienen acceso al estado que es una especie de memoria compartida
def node_1(state: State):
    if state.get("customer_name") is None:
        #actualizamos solo el nombre o las variables especificas que queramos
        return {
            "customer_name": "Jhon Doe"
        }
    return {}
```

```python
from langgraph.graph import StateGraph, START, END

# A este builder le estamos pasando la definición ya establecida en el primer bloque de estado "State"
builder = StateGraph(State)
#nombramos el nodo "node_1" y le decimos quien va a resolver este nodo node_1
builder.add_node("node_1", node_1)

#estas son conexiones, START -> node_1 -> END
builder.add_edge(START, 'node_1')
builder.add_edge('node_1', END)

#compilamos el agente
agent = builder.compile()
```

```python
print(agent.get_graph().draw_ascii())
```
