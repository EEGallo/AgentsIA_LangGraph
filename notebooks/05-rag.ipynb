```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("openai:gpt-4o", temperature=0)
```

```python
from openai import OpenAI
# 1) Cliente oficial de OpenAI para manejo de vector stores

client = OpenAI()

#hacemos una declaración de la TOOLS de OpenIA para nuestro agente
# --- SOLO LA PRIMERA VEZ: crear vector store y asociar archivo ---

# 1) Crear vector store
vs = client.vector_stores.create(name="curso-langgraph")
vector_store_id = vs.id


# 2) Adjuntar el archivo existente al vector store
client.vector_stores.files.create(
    vector_store_id=vector_store_id,
    file_id="file-3LwQZkX4aAumM4dk2HBKsq", #este es el ID del file que se carga en storage de OpenAI
)

print("Vector store:", vector_store_id)


# --- SIEMPRE QUE VAYAS A USAR EL LLM CON FILE_SEARCH ---


# Definición del tool en formato OpenAI
file_search_tool = {
    "type": "file_search",
    "vector_store_ids": [vector_store_id],
}

# Bind del modelo con ese tool
llm_with_tools = llm.bind_tools([file_search_tool])


from langchain_core.messages import HumanMessage


messages = [
    HumanMessage(content="¿Qué dice el documento sobre LangGraph?")
]


resp = llm_with_tools.invoke(messages)
resp.text
```
