```python
from langchain.chat_models import init_chat_model
llm = init_chat_model("openai:gpt-4o", temperature=0)

response = llm.invoke('Hola soy Ema y quiero que me cuentes un chiste')
response.text
```

```python
from pydantic import BaseModel, Field
from langchain_core.messages import SystemMessage, HumanMessage


class ContactInfo(BaseModel):
    """Contact information for a person."""

    name: str | None = Field(default=None, description="The name of the person")
    email: str | None = Field(default=None, description="The email address of the person")
    phone: str | None = Field(default=None, description="The phone number of the person")
    tone: int | None = Field(
        default=None,
        description="Tone from 0 (very negative) to 100 (very positive)",
        ge=0,
        le=100,
    )
    age: int | None = Field(default=None, description="The age of the person in years")
    sentiment: str | None = Field(
        default=None,
        description="Overall sentiment of the conversation (e.g., 'positivo', 'neutral', 'negativo')",
    )

llm_with_structured_output = llm.with_structured_output(schema=ContactInfo)

# 3) Mensajes de ejemplo
messages = [
    SystemMessage(
        content="Eres un asistente de ayuda que puede extraer informaci√≥n de contacto desde una conversaci√≥n."
    ),
    HumanMessage(
        content="Hola soy Emanuel y mi telefono es 262445448"
    ),
]

response = llm_with_structured_output.invoke(messages)
response.name
```
